```python
from wordcloud import WordCloud,ImageColorGenerator
import jieba
import matplotlib.pyplot as plt
from imageio import imread
```

```python
f = open("The_Gift_of_the_Magi.txt","r",encoding='utf-8')
t = f.read()
print(t)
# f.close()
```

```python
# 设置词云对象
w = WordCloud(width = 2000,height=1000,background_color="white")
```

```python
# 加载文本
w.generate(t)
```

```python
plt.figure(figsize=(22,10))
#绘制词云图片
plt.imshow(w)
#显示图片时不显示坐标尺寸
plt.axis('off')
#显示词云图片
plt.show()
```

```python
w.to_file("gift.png")
```

```python
#读入背景图片
pic = imread('bg1.jpg')
#生成词云图片
w = WordCloud(
mask=pic, # 画布
background_color='white', # 背景颜色白
margin=1, # 设置图片的边缘
scale=1.5, # 1.5倍
max_font_size=100, # 显示的最大的字体大小
).generate(t)

#产生背景图片，基于彩色图像的颜色生成器
# bg = ImageColorGenerator(pic)           
```

```python
plt.figure(figsize=(22,10))
#绘制词云图片
plt.imshow(w)
# plt.imshow(w.recolor(color_func=bg))

#显示图片时不显示坐标尺寸
plt.axis('off')
#显示词云图片
plt.show()
```

```python
from pyecharts import options as opts
from pyecharts.charts import Bar, Page
from pyecharts.globals import ThemeType
```

### 柱状图

```python
bar = (
    Bar({"theme": ThemeType.MACARONS}) # opts.InitOpts（theme标图主题 = ThemeType.MACARONS）
    .add_xaxis(["氪金", "杂物", "通勤", "学习", "游玩", "花呗", "网购"])
    .add_yaxis("张三", [500, 100, 300, 500, 100, 105, 200])
    .add_yaxis("李四", [1000, 250, 405, 800, 200, 205, 100])
    .add_yaxis("王二麻子", [100, 500, 400, 100, 2000, 25, 180])
    .set_global_opts(title_opts={"text": "钱都去哪了？"})
    )

htl = 'bar.html'
fhtml="file://"+bar.render(htl)
fhtml
```

```python
import webbrowser
```

```python
wb=webbrowser.get('safari')
wb.open_new(fhtml)
```

### 折线图

```python
from pyecharts.charts import Line
```

```python
attr = ["10.13", "10.14", "10.15", "10.16" , "10.17" , "10.18"]
v1 = [1650, 1700, 1461, 1350, 1100, 1500]
v2 = [1020, 575, 400, 350, 330, 480]

line = (
        Line()
        .add_xaxis(attr)
        .add_yaxis("武汉->北京", v1)
        .add_yaxis("武汉->昆明", v2)
        .set_global_opts(title_opts=opts.TitleOpts(title="航班价格折线图"))
       )

htl="line.html"
fhtml="file://"+line.render(htl)
fhtml
```

```python
wb=webbrowser.get('safari')
wb.open_new(fhtml)
```

### 饼统计图

```python
from pyecharts.charts import Pie
```

```python
pie = (
        Pie()
        .add("", [list(z) for z in zip(['上课','睡眠','餐饮','娱乐','学习','写作业'], [2,4,1,1,8,20])])
        .set_colors(["blue", "green", "black", "red", "pink","orange"])
        .set_global_opts(title_opts=opts.TitleOpts(title="作业写完了吗？"))
        .set_series_opts(label_opts=opts.LabelOpts(formatter="{b}: {c}")) 
        # {a}（系列名称），{b}（数据项名称），{c}（数值）, {d}（百分比）
      )

htl="pie.html"
fhtml="file://"+pie.render(htl)
fhtml
```

```python
wb=webbrowser.get('safari')
wb.open_new(fhtml)
```
