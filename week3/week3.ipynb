```python
def triangles():
    L=[1]
    while True:
        yield L #Ëø≠‰ª£Âô®ÔºåÂáΩÊï∞ËøîÂõûÊüê‰∏™ÂÄºÊó∂Ôºå‰ºöÂÅúÁïôÂú®Êüê‰∏™‰ΩçÁΩÆÔºåËøîÂõûÂáΩÊï∞ÂÄºÂêéÔºå‰ºöÂú®ÂâçÈù¢ÂÅúÁïôÁöÑ‰ΩçÁΩÆÁªßÁª≠ÊâßË°åÔºåÁõ¥Âà∞Á®ãÂ∫èÁªìÊùü„ÄÇ
        L=[sum(i) for i in zip([0]+L,L+[0])]

a=triangles()
print(a)
for i in range(7):
    ss=" ".join([str(d) for d in next(a)])
    print("{0:^25}".format(ss)) # ^Â±Ö‰∏≠
```
